```python
w = df[df["Ship Mode"]=="First Class"]
x = df[df["Ship Mode"]=="Second Class"]
y = df[df["Ship Mode"]=="Standard Class"]
z = df[df["Ship Mode"]=="Same Day"]
```

```python
n = np.array([np.average(w["Profit"]), np.average(x["Profit"]), np.average(y["Profit"]), np.average(z["Profit"])])
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = '#0018ed'
plt.rcParams['axes.labelcolor']= '#FFFF00'
plt.rcParams['xtick.color'] = '#FFFF00'
plt.rcParams['ytick.color'] = '#FFFF00'
plt.rcParams['font.size']=20
mylabels = ["First Class (31.83%)", "Second Class (29.53%)", "Standard Class (27.49%)", "Same Day (29.26%)"]
plt.figure(figsize=(20,10))
wp = { 'linewidth' : 1, 'edgecolor' : "white" } 
plt.pie(n, labels = mylabels, shadow = True, wedgeprops = wp, explode=(0.1, 0.1, 0.1, 0.1) )
plt.legend(title = "Average Sales: ", fontsize=15, facecolor="#b8a5b8", edgecolor="#ffffff", loc ="center right")
plt.show() 
```

```python
lae=LabelEncoder()
df["Ship Mode"] = lae.fit_transform(df["Ship Mode"].astype(str))
df["Segment"] = lae.fit_transform(df["Segment"].astype(str))
df["City"] = lae.fit_transform(df["City"].astype(str))
df["State"] = lae.fit_transform(df["State"].astype(str))
df["Region"] = lae.fit_transform(df["Region"].astype(str))
df["Category"] = lae.fit_transform(df["Category"].astype(str))
df["Sub-Category"] = lae.fit_transform(df["Sub-Category"].astype(str))
corelation = df.corr()
plt.subplots(figsize=(20,15))
_ = sns.heatmap(corelation,xticklabels=corelation.columns, yticklabels=corelation.columns,annot=True)
```

First class products are sold more than Second class, Standard class and Same day.
So we should give discount more to other classes to increase the profit.
